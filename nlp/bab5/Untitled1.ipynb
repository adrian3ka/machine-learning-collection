```python
import nltk.tag
from nltk.tokenize import sent_tokenize, word_tokenize
from sklearn.externals import joblib
import os.path
```

```python
sent_tagged = []

dirpath = os.getcwd()
fileName = [dirpath + "/./UI-1M-tagged.txt"]

print fileName
for f in fileName:
    print "Learning >> " + f
    file = open(f, "r")
    sent = file.read()
    data_sents = [nltk.tag.str2tuple(t) for t in sent.split()]
    temp_sent = []
    stop_word = [".", "!", "?"]
    for data in data_sents:
        temp_sent.append(data)
        if data[0] in stop_word:
            sent_tagged.append(temp_sent)
            temp_sent = []
```

```python
print sent_tagged[0]
size = int(len(sent_tagged) * 0.99)
train_sents = sent_tagged[:size]
test_sents = sent_tagged[size:]
```

```python
default_tagger = nltk.DefaultTagger('NN')
default_tagger.evaluate(test_sents)
```

```python
unigram_tagger = nltk.UnigramTagger(train_sents, backoff=default_tagger)
unigram_tagger.evaluate(test_sents)
```

```python
bigram_tagger = nltk.BigramTagger(train_sents, backoff=default_tagger)
bigram_tagger.evaluate(test_sents)
```
