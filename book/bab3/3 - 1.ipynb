### Dealing with HTML

```python
url = "http://news.bbc.co.uk/2/hi/health/2284783.stm"
```

```python
html_doc = urlopen(url).read()
```

```python
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
raw = soup.get_text()
```

```python
tokens = nltk.word_tokenize(raw)
```

```python
tokens
```

```python
text = nltk.Text(tokens)
text.concordance('gene')
```
