```python
def vowel_count(word):
    counts = 0
    for w in word:
        if ( w == 'a' or w == 'i' or w == 'u' or w == 'e' or w == 'o'):
            counts += 1
    return counts
```

```python
def my_gender_features(word):
    vowel_counts = vowel_count(word)
    return {'last_letter': word[-1], 'vowel_counts': vowel_counts}
```

```python
featuresets = [(my_gender_features(n), g) for (n,g) in names]
train_set, test_set = featuresets[500:], featuresets[:500]
my_classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
print nltk.classify.accuracy(my_classifier, test_set)
```

```python
my_classifier.show_most_informative_features(99)
```

## Choosing the Right Features

```python
def gender_features2(name):
    features = {}
    features["firstletter"] = name[0].lower()
    features["lastletter"] = name[-1].lower()
    for letter in 'abcdefghijklmnopqrstuvwxyz':
        features["count(%s)" % letter] = name.lower().count(letter)
        features["has(%s)" % letter] = (letter in name.lower())
    return features
```

## Document Classification

```python
from nltk.corpus import movie_reviews
documents = [(list(movie_reviews.words(fileid)), category)
             for category in movie_reviews.categories()
             for fileid in movie_reviews.fileids(category)]
random.shuffle(documents)
```

```python
all_words = nltk.FreqDist(w.lower() for w in movie_reviews.words())
word_features = all_words.keys()[:2000]
```

```python
def document_features(document):
    document_words = set(document)
    features = {}
    for word in word_features:
        features['contains(%s)' % word] = (word in document_words)
    return features
```

```python
print document_features(movie_reviews.words('pos/cv957_8737.txt'))
```

```python
featuresets = [(document_features(d), c) for (d,c) in documents]
train_set, test_set = featuresets[100:], featuresets[:100]
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
print nltk.classify.accuracy(classifier, test_set)
```

```python
classifier.show_most_informative_features(5)
```

## Part-of-Speech (POS) Tagging

```python
from nltk.corpus import brown
from operator import itemgetter

suffix_fdist = nltk.FreqDist()
for word in brown.words():
    word = word.lower()
    suffix_fdist[word[-1:]] += 1
    suffix_fdist[word[-2:]] += 1
    suffix_fdist[word[-3:]] += 1
```

```python
common_suffixes = sorted(suffix_fdist.items(), key=itemgetter(1), reverse=True)[:100]
```

```python
print common_suffixes
```

```python
def pos_features(word):
    features = {}
    for suffix,count in common_suffixes:
        features['endswith(%s)' % suffix] = word.lower().endswith(suffix)
    return features
```

```python
tagged_words = brown.tagged_words(categories='news')
featuresets = [(pos_features(n), g) for (n,g) in tagged_words]
```

```python
size = int(len(featuresets) * 0.1)
train_set, test_set = featuresets[size:], featuresets[:size]
```

```python
classifier = nltk.DecisionTreeClassifier.train(train_set)
nltk.classify.accuracy(classifier, test_set)
```

```python
classifier.classify(pos_features('cats'))
```

## Exploiting Context

```python
def pos_features(sentence, i):
    features = {"suffix(1)": sentence[i][-1:],
                "suffix(2)": sentence[i][-2:],
                "suffix(3)": sentence[i][-3:]}
    if i == 0:
        features["prev-word"] = "<START>"
    else:
        features["prev-word"] = sentence[i-1]
    return features
```

```python
pos_features(brown.sents()[0], 8)
```

```python
tagged_sents = brown.tagged_sents(categories='news')
```

```python
featuresets = []
```

```python
for tagged_sent in tagged_sents:
    untagged_sent = nltk.tag.untag(tagged_sent)
    for i, (word, tag) in enumerate(tagged_sent):
        print i,word,tag
        featuresets.append((pos_features(untagged_sent, i), tag) )
```
